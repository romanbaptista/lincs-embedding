# CANADA: Venn

```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
from matplotlib_venn import venn2_unweighted, venn2
import matplotlib.pyplot as plt
```

```python
OUTPUT = 'D:/DDesktop/_work/data/canada/output/'
```

```python
# Load data
df_result = pd.read_csv(OUTPUT + 'df_biovscds_result.csv')
# Set perturbation to lower
df_result['bio_Perturbation'] = df_result['bio_Perturbation'].str.lower()
df_result['cds_Perturbation'] = df_result['bio_Perturbation'].str.lower()
df_result.head()
```

```python
# Set tags
list_tags = ['BioJupies', 'L1000CDS']

# Venn diagrams
for column_type in ['Perturbation', 'Cell-line', 'Dose', 'Time']:
    
    # Initialise list of sets
    list_sets = []
    
    # Get sets
    set_bio = set(df_result[f'bio_{column_type}'])
    set_cds = set(df_result[f'cds_{column_type}'])
    print(sorted(set_bio))
    print(sorted(set_cds))

    # Append sets
    list_sets.append(set_bio)
    list_sets.append(set_cds)

    # Venn diagram
    plt.title(f'{column_type}')
    venn2_unweighted(list_sets,
                    set_labels = list_tags,
                    alpha = 0.5)

    plt.show()
```
