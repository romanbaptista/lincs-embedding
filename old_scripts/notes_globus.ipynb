# Globus Personal Connect

Full setup instructions for the Linux installation/configuration of Globus Personal Connect (GPC) is found [here](https://docs.globus.org/globus-connect-personal/install/linux/).

Here, the steps taken to install/run the program without a GUI were used.

## Download

After logging into to the cluster, navigate to your user directory (if not there automatically) and create a `globus` directory:

```bash
mkdir globus
```

Navigate to that directory:

```bash
cd globus
```

Once inside your `globus` directory, download the tarball for GPC:

```bash
wget https://downloads.globus.org/globus-connect-personal/linux/stable/globusconnectpersonal-latest.tgz
```

Extract the files from the downloaded tarball:

```bash
tar xzf globusconnectpersonal-latest.tgz
```

This will create a directory within `globus` with a specific version of GPC e.g. `/globus/globusconnectpersonal-x.y.z` where `x.y.z` is a specific version number.

## Setup

Navigate to the version-specific GCP directory, where `x.y.z` is the specific version of the folder:

```bash
cd globusconnectpersonal-x.y.z
```

Begin running GCP, if this is the first time running it, the setup will activate:

```bash
./globusconnectpersonal
```

As there is no visual interface on the HPC, all prompts will be text-based. After following these, a message stating successful setup will appear.

## Running GPC

To run GPC, for example after logging back into the HPC, ensure that you have navigated to the `globuspersonalconnect-x.y.z` directory, and use the following command:

```bash
./globusconnectpersonal -start &
```

In the case of any errors, use the following command for more information:

```bash
./globusconnectpersonal -start -debug
```

You can check the status of GPC using the following command:

```bash
./globusconnectpersonal -status
```
