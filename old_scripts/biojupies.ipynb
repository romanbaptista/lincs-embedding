# BioJupies

```python
import pandas as pd
import numpy as os
import os
```

```python
INPUT = 'D:/DDesktop/_work/data/canada/input/'
FEATURECOUNTS = INPUT + 'featureCounts/'
OUTPUT = 'D:/DDesktop/_work/data/canada/output/'
```

```python
counts = [file for file in os.listdir(FEATURECOUNTS) if 'summary' not in file]

num_files = len(counts)
print('num_files:', num_files)

df_results = pd.DataFrame()

# Get raw HALO 6h files
for file in counts:

    if '_Halo6h' in file:

        # print(file)
        filename = file.split('_')[1]
        # print(filename)

        df = pd.read_csv(FEATURECOUNTS + file, sep = '\t', comment = '#')
        df = df.iloc[:, [0, -1]]
        df.rename(columns = {df.columns[-1] : filename}, inplace = True)
        df_results = pd.concat([df_results, df])
    
    if '_Ve6h' in file:

        filename = file.split('_')[1]
        df = pd.read_csv(FEATURECOUNTS + file, sep = '\t', comment = '#')
        df = df.iloc[:, [0, -1]]
        df.rename(columns = {df.columns[-1] : filename}, inplace = True)
        df_results = pd.concat([df_results, df])


# Melt
df_results = pd.melt(df_results, id_vars = 'Geneid')
# Drop duplicates
df_results = df_results[~df_results['value'].isna()]
df_results.drop_duplicates(subset = ['Geneid', 'variable'], inplace = True)
# Pivot
df_results = pd.pivot(df_results, index = 'Geneid', columns = 'variable', values = 'value')
# Save file
df_results.to_csv(OUTPUT + 'BIOJUPES.csv', index = True)
# Get vehicle control files
df_results.head()
```
